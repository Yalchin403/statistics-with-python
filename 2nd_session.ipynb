```python
q1 = data_frame["horsepower"].quantile(0.25)
q1
q3 = data_frame["horsepower"].quantile(0.75)
q3

iqr = q3 - q1
iqr

upper_boundary = q3 + 1.5 * iqr
lower_boundary = q1 - 1.5 * iqr

upper_boundary
lower_boundary
```

```python
filter = data_frame["horsepower"] < upper_boundary
data_frame2 = data_frame[filter]
data_frame2
```

```python
filter = data_frame2["horsepower"] > lower_boundary
data_frame3 = data_frame2[filter]
data_frame3
```

```python
from random import randrange
import random 
import string


def dict_gen(n, m):
    dict_ = dict()
    for i in range(n):
        key = f"key{i}"
        dict_[key] = []
        for i in range(m):
            letters = string.ascii_lowercase
            result_str = ''.join(random.choice(letters) for i in range(6))
            dict_[key].append(result_str)
            
    return dict_

dict_ = dict_gen(5, 20)
data_frame = pd.DataFrame(dict_)
data_frame
```
